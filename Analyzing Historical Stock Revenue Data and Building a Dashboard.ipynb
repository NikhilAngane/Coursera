```python
import yfinance as yf

# Define the ticker symbol
ticker_symbol = "TSLA"

# Create a Ticker object
tsla_ticker = yf.Ticker(ticker_symbol)

# Fetch historical stock data
tesla_data = tsla_ticker.history(period="max")

# Reset the index
tesla_data_reset = tesla_data.reset_index()

# Save the reset dataframe to a CSV file
tesla_data_reset.to_csv("tesla_data_reset.csv", index=False)

# Display the first five rows of the reset dataframe
print(tesla_data_reset.head())
```

```python
print(tesla_data_reset.tail())
```

```python
import yfinance as yf

# Define the ticker symbol
ticker_symbol = "GME"

# Create a Ticker object
gme_ticker = yf.Ticker(ticker_symbol)

# Fetch historical stock data
gme_data = gme_ticker.history(period="max")

# Reset the index
gme_data_reset = gme_data.reset_index()

# Save the reset dataframe to a CSV file
gme_data_reset.to_csv("gme_data_reset.csv", index=False)

# Display the first five rows of the reset dataframe
print(gme_data_reset.head())
```

```python
print(gme_data_reset.tail())
```

```python
import matplotlib.pyplot as plt

# Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Open'], label='Opening Price')
    plt.plot(data['Date'], data['Close'], label='Closing Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
make_graph(tesla_data_reset, title="Tesla Stock Data") 
```
